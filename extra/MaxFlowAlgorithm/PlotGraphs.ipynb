```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para ler o arquivo e criar o grafo
def read_file(nome_arquivo):
    with open(nome_arquivo, 'r') as arquivo:
        # Cria um grafo direcionado
        G = nx.DiGraph()

        # Lê as linhas sequentes com as conexões e pesos
        for linha in arquivo:
            u, v, cap = linha.split()
            u = int(u)
            v = int(v)
            cap = float(cap)

            # Adiciona as arestas ao grafo com os pesos
            if G.has_edge(u,v):
                G[u][v]['capacity'] += cap
            else:
                G.add_edge(u, v, capacity=cap)

        return G

# Função para plotar o grafo
def plot_graph(G, node_size):
    pos = nx.kamada_kawai_layout(G)  # Define a posição dos vértices
    weights = nx.get_edge_attributes(G, 'capacity')  # Obtém os pesos das arestas

    # Plota os vértices
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=node_size)

    # Plota as arestas
    nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)

    # Plota os rótulos das arestas com os pesos
    nx.draw_networkx_edge_labels(G, pos, edge_labels=weights)

    # Plota os rótulos dos vértices
    nx.draw_networkx_labels(G, pos, font_color='black')

    # Exibe o gráfico
    plt.axis('off')
    plt.show()
```

```python
# Nome do arquivo de texto com o grafo
nome_arquivo = 'TestCase/test1.txt'

# Lê o arquivo e cria o grafo
grafo = read_file(nome_arquivo)

# Plota o grafo
plot_graph(grafo, 500)

# Calcular o fluxo máximo usando o algoritmo de Ford-Fulkerson
flow_value, flow_dict = nx.maximum_flow(grafo, 1, 4)

# Imprimir o valor do fluxo máximo
print("Fluxo máximo:", flow_value)

# Imprimir o fluxo em cada aresta
for u in flow_dict:
    for v in flow_dict[u]:
        print(f"Fluxo de {u} para {v}: {flow_dict[u][v]}")
```
